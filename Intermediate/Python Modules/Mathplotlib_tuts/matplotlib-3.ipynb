```python
import matplotlib as mpl

import matplotlib.pyplot as plt

import numpy as np
```

# Labels

With Pyplot, you can use the xlabel() and ylabel() functions to set a label for the x- and y-axis.



```python

x_axis = np.array([3,5,2,1])
y_axis = np.array([9,8,7,6])

plt.plot(x_axis, y_axis, color = "blue")

# assigning the labels to each axis
plt.xlabel('magnitude')
plt.ylabel('average magnitude')

plt.show()
```

# Title 

With Pyplot, you can use the title() function to set a title for the plot.



```python

x_axis = np.array([3,5,2,1])
y_axis = np.array([9,8,7,6])

plt.plot(x_axis, y_axis, color = "red" )

# adding tile
plt.title('BrokenPipeError')

# assigning the labels to each axis
plt.xlabel('magnitude')
plt.ylabel('average magnitude')

plt.show()
```

# Set Font Properties for Title and Labels

You can use the fontdict parameter in xlabel(), ylabel(), and title() to set font properties for the title and labels.



```python
x_axis = np.array([3,5,2,1,5])
y_axis = np.array([9,8,7,6,7])

plt.plot(x_axis, y_axis, color = "green", linestyle = ':' )

# adding tile
plt.title('BrokenPipeError')

#Adding font properties for title and labels
prop_x = {'family': 'Arial', 'size': 12, 'color': 'blue'}
prop_y = {'family': 'serif', 'size': 12, 'color': 'red'}

# assigning the labels to each axis
plt.xlabel('magnitude', fontdict = prop_x)
plt.ylabel('average magnitude', fontdict = prop_y)

plt.show()
```

# Grid

With Pyplot, you can use the grid() function to add grid lines to the plot.



```python
x_axis = np.array([3,4,2,1])
y_axis = np.array([9,8,7,6])

plt.plot(x_axis, y_axis, color = "red" )

# adding tile
plt.title('BrokenPipeError')

# assigning the labels to each axis
plt.xlabel('magnitude')
plt.ylabel('average magnitude')

plt.grid()

plt.show()
```

# specify the grid lines to display

You can use the axis parameter in the grid() function to specify which grid lines to display.

Legal values are: 'x', 'y', and 'both'. Default value is 'both'.

```python
x_axis = np.array([2,4,6,8])
y_axis = np.array([9,8,7,6])

plt.plot(x_axis, y_axis, color = "k" )

# adding tile
plt.title('BrokenPipeError')

# assigning the labels to each axis
plt.xlabel('magnitude')
plt.ylabel('average magnitude')

plt.grid(axis = 'x')

plt.show()
```

```python
x_axis = np.array([3,4,2,1])
y_axis = np.array([1,9,7,6])

plt.plot(x_axis, y_axis, color = "red" )

# adding tile
plt.title('BrokenPipeError')

# assigning the labels to each axis
plt.xlabel('magnitude')
plt.ylabel('average magnitude')

plt.grid(axis = 'y')

plt.show()
```

# Display Multiple Plots


With the subplot() function you can draw multiple plots in one figure:

The subplot() function takes three arguments that describes the layout of the figure.

The layout is organized in rows and columns, which are represented by the first and second argument.

The third argument represents the index of the current plot.

So, if we want a figure with 2 rows an 1 column (meaning that the two plots will be displayed on top of each other instead of side-by-side), we can write the syntax like this:



```python
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1,2,1)
plt.plot(x, y)
plt.title('ArithmeticError')


#plot 2
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1,2,2)
plt.plot(x, y)
plt.title('buffer_error')

plt.show()
```

```python
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2,1,1)
plt.plot(x, y)

#plot 2
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2,1,2)
plt.plot(x, y)

plt.show()
```

You can draw as many plots you like on one figure, just descibe the number of rows, columns, and the index of the plot.



```python
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2,3,1)
plt.plot(x, y)

#plot 2
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2,3,2)
plt.plot(x, y)

#plot 3:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2,3,3)
plt.plot(x, y)

#plot 4
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2,3,4)
plt.plot(x, y)

plt.show()
```

# Super Title

You can add a title to the entire figure with the suptitle() function:



```python
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1,2,1)
plt.plot(x, y)
plt.title('ArithmeticError')


#plot 2
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1,2,2)
plt.plot(x, y)
plt.title('buffer_error')

plt.suptitle('BrokenPipeError')

plt.show()
```
