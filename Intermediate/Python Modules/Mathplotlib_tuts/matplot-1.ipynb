# Matplot liblary tutorial

```python
# Importing the necessary libraries
import matplotlib as mpl
import matplotlib.pyplot as plt  # Most of the Matplotlib utilities lies under the pyplot submodule, and are usually imported under the plt alias:


```

```python
# Title of the graph
plt.title('BufferError Prediction')


# x axis values
x = [12,34,56,68,110]

# y axis values
y = [110,230,430,460,850]


# Naming the x axis
plt.xlabel('x-axis')

# Naming the y axis
plt.ylabel('y-axis')

#plotting The points 
plt.plot(x,y)

# Function to show the plot
plt.show()
```
