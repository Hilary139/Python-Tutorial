# Line styles
You can use the keyword argument linestyle, or shorter ls, to change the style of the plotted line:



character and meaning

'-'

solid line style

'--'

dashed line style

'-.'

dash-dot line style

':'

dotted line style

```python
x_line_style = np.array([3,4,6,5,7,9,8])

y_line_style = np.array([20,33,87,22,55,44,33])

plt.plot(x_line_style, y_line_style, linestyle = ':') # Or you can use ls = ':'

plt.show()


'b'    # blue markers with default shape
'or'   # red circles
'-g'   # green solid line
'--'   # dashed line with default color
'^k:'  # black triangle_up markers connected by a dotted line
```

# Format Strings fmt

You can use also use the shortcut string notation parameter to specify the marker.

This parameter is also called fmt, and is written with this syntax:

marker|line|color

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, 'o:g')
plt.show()
```

# Marker size 
You can use the keyword argument markersize or the shorter version, ms to set the size of the markers:

```python
xpoint = np.array([3,2,1,6,2,3,4,-5,-4])

plt.plot(xpoint, marker = 'o' , ms = 20 , color ='g')

plt.show()
```

# Marker color (markeredgecolor)

You can use the keyword argument markeredgecolor or the shorter mec to set the color of the edge of the markers:

```python
xpoint = np.array([3,2,1,6,2,3,4,-5,-4])

plt.plot(xpoint, marker = 'o' , ms = 20 , color ='g', mec = 'r')

plt.show()
```

# Marker color (markerfacecolor)

You can use the keyword argument markerfacecolor or the shorter mfc to set the color inside the edge of the markers:

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 20, mfc = 'b', color = 'k')
plt.show()
```

# marker color (using both mec and mfc)

```python
ypoints = np.array([3, 8, 11, 10])

plt.plot(ypoints, marker = 'o', ms = 20, mfc = 'b', color = 'k', mec = 'r')
plt.show()
```
